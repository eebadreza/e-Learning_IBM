```python
dataset = pd.read_csv("dataset/data.csv")
dataset.head()
```

```python
targets = dataset.diagnosis
targets.value_counts()
# targets
```

```python
targets = np.where(targets.values == 'M', 0, 1)
# targets.value_counts()
```

```python
list = ['Unnamed: 32', 'id', 'diagnosis']
# print(list)
data = dataset.drop(list, axis=1)
# print(data)
data.shape
# targets
```

Normalizing Features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# print(data.radius_mean[0])
# x = (data.radius_mean[0] - data.radius_mean.min()) / (data.radius_mean.max() - data.radius_mean.min())
# print(x)
data = scaler.fit_transform(data)
data
```

Splitting data into Training and Test parts

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, targets, test_size=0.1)
X_test.shape, Y_test.shape, X_train.shape, Y_train.shape
```

Features : X 
Targets : Y
Parameter K : Number of Neighbours
KNN using sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors' : [5,7,11,15,3],'algorithm': ['auto','ball_tree','kd_tree','brute'],'leaf_size': [10,30, 50, 100]}
```

```python
grid_search = GridSearchCV(knn, param_grid=param_grid)
```

```python
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_estimator_
# grid_search.best_score_
```

```python
from sklearn.metrics import accuracy_score
predict = grid_search.predict(X_test)
# print(classification_report(y_test, grid_predictions))
accuracy_score(Y_test, predict)
# predict
```
